```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/Users/santiagobermejo/Desktop/IRONHACK/Kaggle-competition/input/training_dataset.csv')
df = df.set_index('id')
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'0': "intercolumnar distance",
'1': "upper margin",
'2': "lower margin",
'3': "exploitation",
'4': "row number",
'5': "modular ratio",
'6': "interlinear spacing",
'7': "weight",
'8': "peak number",
'9': "modular ratio/ interlinear spacing"})
```

```python
df.head()
```

```python
y = df.author
```

```python
X = df.drop(['author'], axis=1)
```

```python
display(X.head())
print(X.dtypes)
print('\n--------------------------\n')
print(X.isnull().sum())
display(X.describe())
display(X.corr())
corr = X.corr()
X.corr()
plt.figure()
sns.heatmap(corr, cmap='YlGnBu')
```

###### MODEL

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

###### RandomForestClassifier

```python
model = RandomForestClassifier(n_estimators=3000, max_depth=50, n_jobs = -1)

model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accurancy', accuracy_score(y_test, y_pred))
```
